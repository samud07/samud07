Popularity Based Recommender System

```python
Ratings_with_name = Ratings_data.merge(Books_data,on='ISBN')
```

```python
Ratings_with_name
```

```python
Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
Numbers_Ratings_df = Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
Numbers_Ratings_df.rename(columns={'Book-Rating':'Numbers_Ratings'},inplace=True)
Numbers_Ratings_df
```

```python
Average_Ratings_df = Ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
Average_Ratings_df.rename(columns={'Book-Rating':'Average_Ratings'},inplace=True)
Average_Ratings_df
```

```python
Popular_df = Numbers_Ratings_df.merge(Average_Ratings_df,on='Book-Title')
Popular_df
```

```python
Popular_df = Popular_df[Popular_df['Numbers_Ratings']>=250].sort_values('Average_Ratings',ascending=False).head(50)
```

```python
Popular_df
```

```python
Popular_df.merge(Books_data,on='Book-Title')
```

```python
Popular_df = Popular_df.merge(Books_data,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Numbers_Ratings','Average_Ratings']]
```
