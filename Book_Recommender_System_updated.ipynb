Popularity Based Recommender System

```python
Ratings_with_name = Ratings_data.merge(Books_data,on='ISBN')
```

```python
Ratings_with_name
```

```python
Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
Numbers_Ratings_df = Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
Numbers_Ratings_df.rename(columns={'Book-Rating':'Numbers_Ratings'},inplace=True)
Numbers_Ratings_df
```

```python
Average_Ratings_df = Ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
Average_Ratings_df.rename(columns={'Book-Rating':'Average_Ratings'},inplace=True)
Average_Ratings_df
```

```python
Popular_df = Numbers_Ratings_df.merge(Average_Ratings_df,on='Book-Title')
Popular_df
```

```python
Popular_df = Popular_df[Popular_df['Numbers_Ratings']>=250].sort_values('Average_Ratings',ascending=False).head(50)
```

Collaborative Filtering Based Recommender System

```python
Ratings_with_name
```

```python
x = Ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200 
padhe_likhe_users = x[x].index
```

```python
filtered_Ratings = Ratings_with_name[Ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_Ratings
```

```python
y = filtered_Ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_Ratings = filtered_Ratings[filtered_Ratings['Book-Title'].isin(famous_books)]
```

```python
final_Ratings
```

```python
pt = final_Ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
similarity_scores
```

```python
def recommend(book_name):
  # index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('1984')
```
